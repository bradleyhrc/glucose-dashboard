```python
import pandas as pd
import numpy as np
import datetime as dt
import panel as pn
import holoviews as hv
pn.extension('tabulator')

import hvplot.pandas
from datetime import timedelta
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/bradleyhrc/glucose-data/main/BHC-Glucose-Data-Clean.csv')
```

```python
pd.DataFrame(df) # looking at the data
```

```python
df.columns # looking at the data
```

```python
df['Device_Timestamp'] = pd.to_datetime(df['Device_Timestamp'], format='%d-%m-%Y %H:%M')
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df['Date'] = df['Date'].dt.date
df
# df['date_col'] =  pd.to_datetime(df['date_col'], format='%d/%m/%Y')
#pd.to_datetime(pd.Series(['05/23/2005']), format="%m/%d/%Y")
```

```python
df.loc[df['Glucose_mmol/L'] > 8.0, 'Range'] = 'High'
df.loc[df['Glucose_mmol/L'] < 3.9, 'Range'] = 'Low'
df.loc[(df['Glucose_mmol/L'] <= 8.0) & (df['Glucose_mmol/L'] >= 3.9), 'Range'] = 'Target'

df.loc[df['Range'] == 'High', 'High'] = df['Glucose_mmol/L'] 
df.loc[df['Range'] == 'Low', 'Low'] = df['Glucose_mmol/L'] 
df.loc[df['Range'] == 'Target', 'Target'] = df['Glucose_mmol/L']

df.loc[df['Glucose_mmol/L'] > 1, 'Region'] = 6

df
```

```python
# data already pre-processed, will add data preprocessing automation here in future update
# make data interactive
idf = df.interactive()
```

```python
#date_picker
date_picker = pn.widgets.DatePicker(name='Date Picker', width=800)
date_picker.end = dt.date(2021,6,5)
date_picker.start = dt.date(2019,8,12)
date_picker.value = dt.date(2021,6,5)
date_picker
```

```python
glucose_pipeline = (
    idf[
        (idf.Date == date_picker) & (idf.Device == 'FreeStyle Libre')
    ]
    .reset_index()
    .sort_values(by='Device_Timestamp')
    .reset_index(drop=True)
)
```

```python
#glucose_pipeline
```

```python
colors = ['#77dd77', 'black', '#f75d59']
dimy = 300

glucose_plot = glucose_pipeline.hvplot(x='Device_Timestamp', 
                                       xlabel='Time', 
                                       y=['Region', 'Glucose_mmol/L', 'Low'], ylim=[0,23],
                                       ylabel='Blood Glucose (Mmol/L)', 
                                       kind='line',
                                       width=800,
                                       height=300,
                                       line_width=[dimy/7.5,3,3], 
                                       title = 'Bradley\'s Glucose Levels',
                                       grid=True,
                                       color=colors,
                                       legend=False,
                                       line_alpha=[0.3,1,1],
                                       hover=False
                                      )
glucose_plot
```

```python
# table pipeline

condensed_idf = df[['Device', 'Device_Timestamp', 'Date', 'Range', 'Glucose_mmol/L']].interactive()

date_picker1 = pn.widgets.DatePicker(name='Date Picker', width=600)
date_picker1.end = dt.date(2021,6,5)
date_picker1.start = dt.date(2019,8,12)
date_picker1.value = dt.date(2021,6,5)

device_picker1 = pn.widgets.RadioButtonGroup(
    name='Device Picker',
    options=['FreeStyle Libre', 'FreeStyle LibreLink'],
    button_type='success',
    width=600
)

glucose_table = (
    condensed_idf[
        (condensed_idf.Date == date_picker1) & (condensed_idf.Device == 'FreeStyle Libre')
    ]
    .sort_values(by='Device_Timestamp')
    .reset_index(drop=True)
    .pipe
)

glucose_table(pn.widgets.Tabulator, pagination='remote', page_size=20, sizing_mode='stretch_both')
```

```python
date_slider = pn.widgets.DateSlider(
    name='Date Slider',
    start=dt.date(2019, 8, 12),
    end=dt.date(2021,6,5),
    value=dt.date(2021,6,5)-dt.timedelta(days=7)
)
date_slider
```

```python

target_idf = df[['Device', 'Device_Timestamp', 'Date', 'Range', 'Glucose_mmol/L', 'High', 'Target', 'Low']].interactive()

time_in_target_pipeline = (
    target_idf[
        (target_idf.Device == 'FreeStyle Libre') & (target_idf.Date >= date_slider)
    ]
    .groupby(['Range']).count()
)

time_in_target_barplot = time_in_target_pipeline.hvplot(kind='barh', 
                                                        x='Range', xlabel='Range', 
                                                        y='Glucose_mmol/L', ylabel='Count',
                                                        title='Time In Target Report'
                                                       )
time_in_target_barplot.opts(color='Range', cmap='Set1')
```
